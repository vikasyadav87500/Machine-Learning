```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('C:/Users/vikas/Downloads/ml-100k/ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))
m_cols = ['movie_id', 'title']
movies = pd.read_csv('C:/Users/vikas/Downloads/ml-100k/ml-100k/u.item', sep='|',encoding = "ISO-8859-1", names=m_cols, usecols=range(2))
```

```python
# merge both the datasets
dataset = movies.merge(ratings , on='movie_id')
```

```python
dataset.head(10)
```

```python
dataset.shape
```

```python
# Creating the pivot_table
userRatings = dataset.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head(10)
```

```python
userRatings.shape
```

```python
# calculating the correlation matrix
corrMatrix = userRatings.corr()
corrMatrix.head(10)
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head(10)
```

```python
corrMatrix.shape
```

```python
print(userRatings.loc[2].dropna().count())
```

```python
print("Movies and their ratings for custom user 1")
#Can change user for 1 to any user and obtain recommendation for that user. Just change loc[number].
myRatings = userRatings.loc[2].dropna()
print('Count of user 2: -' , myRatings.count())
myRatings.head(10)
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print( "adding sims for " + myRatings.index[i] + "...")
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[i])
    simCandidates = simCandidates.append(sims)
    
print( "sorting in decreasing order of similarity score: ")
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
type(corrMatrix[myRatings.index[4]].dropna())
simCandidates.shape
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
simCandidates.head()
```

```python
simCandidates.shape
```

```python
print("Adding up the similarity scores of duplicate values:")
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
#simCandidates.shape
```

```python
# Returns the Top 10 recommendations
print("Filtering the result to remove already rated movies:")
filteredSims = simCandidates.drop(myRatings.index, errors='ignore')
filteredSims.head(10)
```
