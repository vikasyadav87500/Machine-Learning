```python
print("Combining both the datasets on the basis of id.....")
credits_new = credits.rename(index=str, columns={"movie_id": "id"})
dataset = movies.merge(credits_new, on='id')
dataset.head(10)
```

```python
# Data-Cleaning. Removing the  irrelevant information
cleaned_df = dataset.drop(columns=['budget','homepage', 'title_x', 'title_y', 'status','production_countries'])
cleaned_df.head(10)
```

```python
# Filling NaNs with empty string in overview column
cleaned_df['overview'] = cleaned_df['overview'].fillna('')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_vector = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
# Fitting the TF-IDF on the 'overview' text
spare_matrix = tf_vector.fit_transform(cleaned_df['overview'])
```

```python
spare_matrix.shape
```

## Forming sigmoid matrix which gives the similarity between overviews

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(spare_matrix, spare_matrix)
```
