```python
class Univariate_Linear_Regression:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        self.coef = np.random.uniform(low=-1 , high =1)
        self.bias = np.random.random()
        
    def hypothesis(self,x):
        return self.coef*x + self.bias
    
    def apply_gradient_descent(self , learning_rate = 0.1):
        dldw0=0
        dldw1=0
        
        #compute gradients
        for x,y in zip(self.x,self.y):
            yhat = self.hypothesis(x)    # get_hypothesis
            grad_00 = (yhat-y)           
            grad_01 = (yhat-y)*x
            dldw0+=grad_00
            dldw1+=grad_01
            
        dldw1 /= self.x.shape[0]
        dldw0 /= self.x.shape[0]
        
        #updating the values
        self.coef = self.coef - (learning_rate*dldw1)
        self.bias = self.bias - (learning_rate*dldw0)
    
    def compute_loss(self):
        losses = []
        for x,y in zip(self.x , self.y):
            yhat = self.hypothesis(x)
            loss =  (yhat-y)**2
            losses.append(loss)
        
        losses = np.array(losses)
        return losses.sum()/2*self.x.shape[0]  
    
    def get_all_preds(self):
        preds = []
        for i in self.x:
            preds.append(self.hypothesis(i))
        return preds
    
    def plot_best_fit(x, y, preds, title=''):
        plt.figure()
        plt.title(title)
        plt.scatter(x[:, 0], y)
        plt.plot(x[:, 0], preds, 'r')
        plt.show()
    
    def run(self, losses , iterations =1):
        
        for _ in range(iterations):
            losses.append(self.compute_loss())
            self.apply_gradient_descent()
            preds = self.get_all_preds()
            plot_best_fit(self.x,self.y,preds,title=str(_+1)+'st plot')
            print("loss of {} th iteration is :- {}".format(_ +1, self.compute_loss()))
            
        #print(losses)
        return losses
            
```
