```python
# Formation of Transition matrix

T = {}

n = 5   #N_Grams

for i in range (len(data)-n):
    n_grams = data[i:i+n]
    next_state = data[i+n]
    T_context = T.setdefault(n_grams , {})
    T_context[next_state] = T_context.setdefault(next_state , 0)+1
```

```python
T
```

```python
for row in T:
    s = sum(T[row].values())
    for val in T[row]:
        T[row][val] = T[row][val]/s
```

```python
def predict_state(ngrams):
    if T.get(ngrams) is None:
        return ' '
    mapped_gram = T[ngrams]
    mapped_states = list(mapped_gram.keys())
    mapped_probab = list(mapped_gram.values())
    return np.random.choice(mapped_states , p=mapped_probab)
```

```python
predict_state('Dr')
```

```python
def generate(initial=None , size = 1000):
    sentence = ''
    if initial is None:
        initial = int(np.random.random()*(len(data)-n))
        initial = data[initial:initial+n]
        
    sentence+=initial
    for i in range(size):
        predict = predict_state(initial)
        sentence+=predict
        initial = sentence[-n:]
        
    return sentence
```

```python
generate()
```
